```python
import pandas as pd
import seaborn as sns
import itertools
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np
from pyparsing import alphanums
from tensorflow import lite
```

```python
df=pd.read_csv("survey lung cancer.csv")
```

```python
df.head()
```

```python
type(df['LUNG_CANCER'])
```

```python
df['LUNG_CANCER'] = df['LUNG_CANCER'].map(
                   {'YES':2 ,'NO':1})
```

```python
df.head()
```

```python
df = df.replace({2:1 ,1:0})
```

```python
df.head()
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df['GENDER'], prefix="")
# col1=df1['_M']
# col2=df1['_F']
# df.join(col1)
# df.join(col2)
# df.head()
name=list(df.columns.values)
name
name1=['F','M']
name=[name1,name]
namef=list(itertools.chain(*name))
namef
fin=pd.concat([df1, df], axis=1, ignore_index=True)
fin.columns=namef

fin=fin.drop(['GENDER'],axis=1)
fin.head()
```

```python
fin.corr(method="pearson")
```

```python
sc=MinMaxScaler()
fin.head()
```

```python
# df=sc.fit_transform(fin)
# df
# column_l=list(fin.columns.values)
```

```python
fin.head()
```

```python
# final=pd.DataFrame(df,columns=column_l)
final=fin
```

```python
final.head()
```

```python
train_set, test_set=train_test_split(final,test_size=0.2, random_state=87)
print(len(test_set),len(train_set))
```

```python
train_x=train_set.drop("LUNG_CANCER",axis=1)
train_y=train_set["LUNG_CANCER"].copy()
test_x=test_set.drop("LUNG_CANCER",axis=1)
test_y=test_set["LUNG_CANCER"].copy()
```

```python
train_x
train_y
test_x
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,mean_squared_error
classifier= LogisticRegression(random_state=0)  
classifier.fit(train_x,train_y)

pred_y= classifier.predict(test_x)

accuracy_score(test_y,pred_y)

test_y

pred_y

mse=mean_squared_error(pred_y,test_y)
rmse=np.sqrt(mse)
print(mse)
print(rmse)
```

# SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear']} 
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(train_x, train_y)
```

```python
print(grid.best_params_)
```

```python
model=SVC(kernel='linear',C=10,gamma=1)
```

```python
# print(grid.best_params_)
# pred_y= grid.predict(test_x)
model.fit(train_x,train_y) 
pred_y=model.predict(test_x);
```

```python
accuracy_score(test_y,pred_y)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,train_x,train_y,cv=10)
scores
```

```python
np.mean(scores)
```

```python
from sklearn.model_selection import cross_val_predict
scores=cross_val_predict(model,test_x,test_y,cv=10)
scores
```
